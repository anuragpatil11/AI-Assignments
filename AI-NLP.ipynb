```python
from nltk import sent_tokenize, word_tokenize
```

```python
from nltk.corpus import stopwords
```

```python
stopwords = stopwords.words("english")
```

```python
sent = "I will walk 500 miles and I would walk 500 more , just to be the man who walks a thousand miles to fall down at your door".lower()
```

```python
sent
```

```python
words= word_tokenize(sent)
print(words)
```

```python
sentences = sent_tokenize(sent)
print(sentences)
```

```python
print(stopwords)
```

```python
for word in words:
  if word in stopwords:
    words.remove(word)
print(words)
```

transform into lowercase then print stop words

```python
from nltk import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
print(words)
```

```python
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words)
```

```python
from nltk.stem import SnowballStemmer
```

```python
from collections import Counter
from nltk import pos_tag
```

counting frequency for stemmed words

```python
nltk.download('averaged_perceptron_tagger')
```

```python
count_dict = {}
```

```python
for word in stemmed_words:
  if word in count_dict:
    count_dict[word]+=1
  else:
    count_dict[word]=1
print(count_dict)
```

```python
pos_tagged = pos_tag(stemmed_words)
print(pos_tagged)
```

```python
# counting frequency for part of speech
count = Counter(tag for _, tag in pos_tagged)
print(count)
```
